```python
import numpy as np
import pandas as pd
import datetime as dt
from datetime import timedelta
```

```python
stock_data_df = pd.read_csv('data_organized3.csv')
stock_data_df
```

```python
stock_data_df= stock_data_df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
stock_data_df
```

```python
stock_data_df.info()
```

```python
stock_data_df['Date']=pd.to_datetime(stock_data_df["Date"])
stock_data_df
```

```python
stock_data_df.info()
```

```python
stock_data_df['price'].replace('', np.nan, inplace=True)
```

```python
stock_data_df.dropna(subset=['price'], inplace=True)
```

```python
stock_data_df.info()
```

```python
stock_data_df['ticker'].value_counts()
```

```python
hotels_stock_df = stock_data_df[(stock_data_df['ticker'].isin(['HLT','MAR', 'WH']))]
hotels_stock_df.reset_index(drop=True, inplace=True)
hotels_stock_df
```

```python
hotels_stock_df.info()
```

```python
cruise_stock_df = stock_data_df[(stock_data_df['ticker'].isin(['CCL','RCL']))]
cruise_stock_df.reset_index(drop=True, inplace=True)
cruise_stock_df
```

```python
cruise_stock_df.info()
```

```python
airlines_stock_df = stock_data_df[(stock_data_df['ticker'].isin(['DAL','UAL']))]
airlines_stock_df.reset_index(drop=True, inplace=True)
airlines_stock_df
```

```python
restaurants_stock_df = stock_data_df[(stock_data_df['ticker'].isin(['CAKE','RUTH']))]
restaurants_stock_df.reset_index(drop=True, inplace=True)
restaurants_stock_df
```

```python
retail_stock_df = stock_data_df[(stock_data_df['ticker'].isin(['AMZN','WMT']))]
retail_stock_df.reset_index(drop=True, inplace=True)
retail_stock_df
```

```python
stock_data_df.to_csv('Resources/AllStocks.csv')
```
